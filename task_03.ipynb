```python
import pandas as pd
data = pd.read_excel('Intern_Performance_200_Rows.xlsx')
data.head()
```

```python
x = data[['Attendance_Percent','Tasks_Completed','Feedback_Score']]
y = data['Completed']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
import pandas as pd

new_intern = pd.DataFrame([[85, 8, 4.5]], 
                          columns=['Attendance_Percent','Tasks_Completed','Feedback_Score'])

result = model.predict(new_intern)
print(result)
```
